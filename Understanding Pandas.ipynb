# Understanding how to combine the DataFrames Together

```python
data = pd.DataFrame({"A": [1,2,3], 
                        "B": [4,5,6], 
                        "C": [7,8,9]}, index = ["a","b","c"])
```

```python
data
```

```python
data2 = pd.DataFrame({"A": [11,12,13], 
                        "B": [14,15,16], 
                        "C": [17,18,19]}, index = ["a","b","c"])
data2
```

```python
data3 = pd.DataFrame({"A": [11,12,13], 
                        "B": [14,15,16], 
                        "C": [17,18,19]}, index = ["d","e","f"])
data3
```

```python
data4 = pd.DataFrame({"A": [21,22,23], 
                        "B": [24,25,26], 
                        "D": [27,28,29]}, index = ["a","b","c"])
data4
```

```python
# Basic concatenation with the same index and same Column names - as in most of the cases
pd.concat([data, data2],ignore_index= True)
```

```python
# Concatenation with same header and different index

pd.concat([data, data3]) # since here the columns are same, the concatenation happened nicely
```

```python
pd.concat([data, data3], axis = 0, 
          keys = ["A","B"])
```

```python
# Data Merging

Student = pd.DataFrame({"Name": ["Abhi","Meera", "Seema", "Reeta"], 
                        "Class": [2,3,5,6], 
                        "Subject": ["Math", "Science", "English", "Math"]})
```

```python
Marks = pd.DataFrame({"Name": ["Abhi","Meera", "Seema", "Reeta", "Abhi","Meera", "Seema", "Reeta", "Abhi","Meera", "Seema", "Reeta"], 
                        "Subject": ["Math","Math","Math","Math" ,"Science","Science","Science","Science", "English","English","English","English"], 
                     "Marks": [20, 22, 21, 19, 24, 23, 19, 22, 19, 20, 21, 20]})
```

```python
Student
```

```python
Marks
```

```python
pd.merge(Student,Marks)
```
