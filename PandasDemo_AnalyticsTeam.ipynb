```python
import pandas as pd
```

```python
# Importing the raw data
survey = pd.read_csv("surveydata.csv", sep = "\t")
name = pd.read_csv("Names.csv", sep = "\t", index_col= "Survey Name")
```

```python
#Checking the import of the raw data
survey.head()
```

```python
#checking the structure of the imported Data
survey.info()
```

```python
# Check the Sanity of COlumns names - quite possible that there could be white spaces while we import
survey.columns
```

```python
#Cleaning the columns names so that we can access the column names easily
survey.columns = survey.columns.str.strip()
```

```python
# Lets Clean the Data:
# No Blank Rows required in the Agent Columns
Filter_AgentBlank = survey["calldata.RTargetAgentSelected"].notnull()
t = survey[Filter_AgentBlank]
t
```

```python
#Creating the Actual filter for the Data. 
Filter_AgentBlank = survey["calldata.RTargetAgentSelected"].notnull()
Filter_Q1Blank = survey["survey.iQ1"].notnull()
survey = survey[Filter_AgentBlank & Filter_Q1Blank]
survey
```

```python
survey["totalscore"] = survey[["survey.iQ1", "survey.iQ2", "survey.iQ3"]].fillna(0).sum(axis = 1)
survey["SurveyAnswered"] = survey[["survey.iQ1", "survey.iQ2", "survey.iQ3"]].notnull().sum(axis = 1)
survey
```

```python
SurveyCal = survey.groupby(["calldata.RTargetAgentSelected"])[["totalscore", "SurveyAnswered"]].sum()
```

```python
SurveyCal
```

```python
SurveyCal["SurveyScore"] = SurveyCal["totalscore"]/SurveyCal["SurveyAnswered"]
```

```python
SurveyCal
```

```python
name
```

```python
pd.merge(SurveyCal,name, left_index= True , right_index= True)[["Agent Name", "SurveyScore"]]
```

```python
Final_Data = pd.merge(SurveyCal,name, left_index= True , right_index= True)[["Agent Name", "SurveyScore"]]
```

```python
Final_Data.to_csv( "Results", index=False)
```
